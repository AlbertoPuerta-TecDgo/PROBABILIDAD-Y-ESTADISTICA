```python
import random
```

```python
#La estadistica descriptiva se usa para describir y resumir los valores y las
#observaciones de un conjunto de datos.
#Un ejemplo de ello es una aplicacion de seguimiento del estado fsico registr
#los datos de una persona y su frecuencia cardiaca durante un periodo de diez das

#Creamos una lista que representa la edad de 100 personas
edades = []

#Hacemos un ciclo para generar 100 numeros aleatorios que representan la edad de las personas
for i in range(100):
  #generamos un numero aletorio con la funcion de randint y lo guardamos en la variable
  #numero_aletorio
  numero_aleatorio = random.randint( 18, 70)
  #Agregamos el numeor aleartorio a la lista de valores.
  edades.append(numero_aleatorio)


print(edades)
```

```python
import pandas as pd
```

```python
#Crear un DataFrame de pandas con los numeros aletorios
df = pd.DataFrame(edades, columns =['Edades'])
df.head()
```

```python
#Calcular una tabla de distribucion de frecuencias.
tabla_frecuencias = df['Edades'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edades', 'Frecuencias']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edades')

#Imprimir la tabla de distribucion de frecuencias
print("Tabla de distribucio de frecuencias")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencias'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencias']/ 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencias'].cumsum()
tabla_frecuencias
```

#Regla de sturges.
###1.-Es una regla que sirve para calcular el nmero de clases o intervalos idneio en los que se debe de dividir un conjunto de datos.
###2.-Estable que el nmerod de clases es igual a uno ms el logartmo en base dos del nmero total de datos.
c = 1 +log2(N)
dodnde N es el tamao de la muestra.

```python
#Importamos la libreria math de funciones matemticas
import math
```

```python
#Volvemos a usar la kista de edades
df = pd.DataFrame({'Edades': edades})
df
```

```python
#Calcular el nmero de intervalos utilizando la regla de Sturges.
n = len(edades)
K = int ( 1 + math.log2(100))
print("Valor de n = ", n ,"\nNmero de intervalos =",K)
```

```python
#Crear tabla de frecuencia

tabla_frecuencia = pd.cut (df['Edades'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
#Ordenar la tabla de frecuenciapor el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
#Generar las columnas siguientes
tabla_frecuencia['Frec Rel'] = tabla_frecuencia['Frec'] / n
tabla_frecuencia['Frec Porc'] = tabla_frecuencia['Frec'] / n * 100
tabla_frecuencia['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac'] = tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac'] = tabla_frecuencia['Frec Porc'].cumsum()

#Mostrar la tabla de frecuencias
print(tabla_frecuencia)
```
